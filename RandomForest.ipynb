```python
import pandas as pd
Cumulative_df = pd.read_csv("cumulative.csv")
```

```python
Cumulative_df 
```

```python
Cumulative_df= Cumulative_df.drop(columns=['rowid','kepid','kepoi_name','kepler_name','koi_tce_delivname','koi_fpflag_nt','koi_fpflag_ss','koi_fpflag_co','koi_fpflag_ec'])
Cumulative_df = Cumulative_df.dropna(axis='columns', how='all')
Cumulative_df = Cumulative_df.dropna()
```

```python
y = Cumulative_df['koi_disposition']
from sklearn.preprocessing import LabelEncoder

# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
```

```python
encoded_y
```

```python
X= Cumulative_df.drop(columns=['koi_disposition','koi_pdisposition'])
X 
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Create a random forest classifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X, encoded_y)
rf.score(X, encoded_y)
```

```python
# Random Forests in sklearn will automatically calculate feature importance
importances = rf.feature_importances_
importances
```

```python
# We can sort the features by their importance
sorted(zip(rf.feature_importances_, X.columns), reverse=True)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
# Generate 4 clusters of random data.
from sklearn.datasets.samples_generator import make_blobs

X_data,y = make_blobs(n_samples=300, centers=3,
                     cluster_std=0.60, random_state=0)
```

```python
# Plot the data
plt.scatter(X_data[:, 0], X_data[:, 1])
```

```python
# Use n_clusters=4 as the k value
# We can see from the plot above that there are 4 clusters
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
```

```python
# Fit the model to the data
kmeans.fit(X_data)
```

```python
# Predict the clusters
predicted_clusters = kmeans.predict(X_data)
predicted_clusters
```

```python
# Plot the predicted clusters to see if the model predicted the correct clusters
# This is visual validation that the model was trained correctly.
plt.scatter(X_data[:, 0], X_data[:, 1], c=predicted_clusters, s=50, cmap='viridis')
```
